```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("star_data.txt", skiprows = 1)
```

```python
print(np.shape(data)) # 2642 rows
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]
```

```python
max_mag = np.argmax(magnitude)
min_mag = np.argmin(magnitude)

max_temp = np.argmax(temperature)
min_temp = np.argmin(temperature)
```

```python
print(f"Star {max_mag + 1} with {magnitude[max_mag]} has the largest magnitude.")
print(f"Star {min_mag + 1} with {magnitude[min_mag]} has the lowest magnitude.")
print(f"Star {max_temp + 1} with {magnitude[max_temp]} has the largest temperature.")
print(f"Star {min_mag + 1} with {magnitude[min_temp]} has the lowest temperature.")
```

```python
print(f"Star {min_mag + 1} is the brightest.")
print(f"Star {max_mag + 1} is the dimmest.")
```

```python
log_temperature = np.log(temperature)
log_magnitude = np.log(magnitude)

plt.figure(figsize=(8, 6))

plt.scatter(log_temperature, log_magnitude, label="Stars")
plt.scatter(np.log(temperature[max_mag]), np.log(magnitude[max_mag]), color="black", marker='*', s=300, label="Dimmest Star")
plt.scatter(np.log(temperature[min_mag]), np.log(magnitude[min_mag]), color="purple", marker='*', s=300, label="Brightest Star")
plt.scatter(np.log(temperature[max_temp]), np.log(magnitude[max_temp]), color="green", marker='*', s=300, label="Hottest Star")
plt.scatter(np.log(temperature[min_temp]), np.log(magnitude[min_temp]), color="orange", marker='*', s=300, label="Coldest Star")

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.legend()
plt.title("Hertzsprung-Russell Diagram")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")

plt.show()
```

```python
# If they did the "bonus", it should look like this:
plt.figure(figsize=(8, 6))

sc = plt.scatter(log_temperature, log_magnitude, c=temperature, cmap="plasma", label="Stars")
plt.scatter(np.log(temperature[max_mag]), np.log(magnitude[max_mag]), color="black", marker='*', s=300, label="Brightest Star")
plt.scatter(np.log(temperature[min_mag]), np.log(magnitude[min_mag]), color="purple", marker='*', s=300, label="Dimmest Star")
plt.scatter(np.log(temperature[max_temp]), np.log(magnitude[max_temp]), color="green", marker='*', s=300, label="Hottest Star")
plt.scatter(np.log(temperature[min_temp]), np.log(magnitude[min_temp]), color="orange", marker='*', s=300, label="Coldest Star")

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.colorbar(sc, label="Temperature (K)")  # Add colorbar for temperature
plt.legend()
plt.title("Hertzsprung-Russell Diagram (Colored by Temperature)")
plt.xlabel("Temperature (log K)")
plt.ylabel("Magnitude")
plt.show()
```
