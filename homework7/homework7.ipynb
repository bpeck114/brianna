```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    y = A * np.sin(w * x)
    return y
```

```python
x = np.linspace(0, 2*np.pi, 100)
```

```python
amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(25, 5))

for amp, freq in zip(amplitude, frequency):
    ax[0].plot(x, make_sine_wave(x, amp, 1))
    
    ax[1].plot(x, make_sine_wave(x, 1, freq))
    
    ax[2].plot(x, make_sine_wave(x, amp, freq))
    
ax[0].set_title("Sine Wave: \nVarying Amplitude", fontsize=25)
ax[0].set_xlabel("x values", fontsize=18)
ax[0].set_ylabel("Sine(x)", fontsize=18)

ax[1].set_title("Sine Wave: \nVarying Frequency", fontsize=25)
ax[1].set_xlabel("x values", fontsize=18)
ax[1].set_ylabel("Sine(x)", fontsize=18)

ax[2].set_title("Sine Wave: \nVarying Amplitude and Frequency", fontsize=25)
ax[2].set_xlabel("x values", fontsize=18)
ax[2].set_ylabel("Sine(x)", fontsize=18)

plt.show()
```

```python
# If they did the "bonus", it should look like this:

fig, ax = plt.subplots(1, 3, figsize=(25, 5))

linestyles = ['-', '--', '-.', ':']

for i, amp in enumerate(amplitude):
    ls = linestyles[i % len(linestyles)]
    y = make_sine_wave(x, amp, 1)
    ax[0].plot(x, y, linestyle=ls, label=f"y = {amp} * sin(x * 1)")
    
for i, freq in enumerate(frequency):
    ls = linestyles[i % len(linestyles)]
    y = make_sine_wave(x, 1, freq)
    ax[1].plot(x, y, linestyle=ls, label=f"y = 1 * sin(x * {freq})")
    
for i, (amp, freq) in enumerate(zip(amplitude, frequency)):
    ls = linestyles[i % len(linestyles)]
    y = make_sine_wave(x, amp, freq)
    ax[2].plot(x, y, linestyle=ls, label=f"y = {amp} * sin(x * {freq})")
    
titles = [
    "Sine Wave: \nVarying Amplitude",
    "Sine Wave: \nVarying Frequency",
    "Sine Wave: \nVarying Amplitude and Frequency"
]

for i in range(3):
    ax[i].set_title(titles[i], fontsize=25)
    ax[i].set_xlabel("x values", fontsize=18)
    ax[i].set_ylabel("Sine(x)", fontsize=18)
    ax[i].legend(fontsize=14)

plt.show()
```
